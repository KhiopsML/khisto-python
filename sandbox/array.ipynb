```python
import os

os.environ["KHISTO_BIN_DIR"] = (
    "/home/elouen/python/khisto-python/src/khiops/build/linux-gcc-debug/bin/khisto"
)
from khisto import histogram, histogram_series, histogram_bin_edges
```

```python
import random
import numpy as np

x = np.array([random.gauss(0, 1) for _ in range(1000)])
```

```python
histogram(x)
```

```python
import pyarrow as pa

x = pa.array([random.gauss(0, 1) for _ in range(1000)])
histogram(x, granularity="best", cumulative=True)
```

```python
import jax

x = jax.numpy.array([random.gauss(0, 1) for _ in range(1000)])
```

```python
histogram(x)
```

```python
histogram(x.tolist())
```

```python
import pandas as pd

x = pd.Series([random.gauss(0, 1) for _ in range(1000)])
histogram(x)
```

```python
import polars as pl

x = pl.Series("x", [random.gauss(0, 1) for _ in range(1000)])
histogram(x)
```

```python
histogram_bin_edges(x)
```

```python
x = pd.Series([random.gauss(0, 1) for _ in range(1000)])

histogram_series(x).to_native()
```

```python
x = pl.DataFrame({"x": [random.gauss(0, 1) for _ in range(1000)]})["x"]
x = pl.concat([x, pl.Series("y", [random.gauss(2, 1) for _ in range(1000)])])

histogram_series(x, cumulative=True).to_native()
```
