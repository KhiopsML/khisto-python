## Example 1: Simple Histogram with Automatic Binning

The key feature of Khisto is automatic optimal binning. When you don't specify `nbins`, Khisto analyzes your data and determines the best bin configuration.

```python
import pandas as pd
import plotly.express as px

data = np.random.normal(0, 1, 1000)
data = pd.DataFrame(
    {
        "value": data,
        "category": np.random.choice(["A", "B"], size=1000),
        "pattern": np.random.choice(["X", "Y"], size=1000),
    }
)

fig = histogram(
    data,
    x="value",
    orientation="h",
    color="category",
    pattern_shape="pattern",
    # facet_col="pattern",
    title="Histogram with Khiops Series",
    color_discrete_sequence=px.colors.qualitative.Set2,
    facet_row="category",
    barmode="group",
    marginal="rug",
)
fig.show()
```

```python
# Generate normal distribution data
data = np.random.normal(0, 1, 1000)

# Create histogram with automatic optimal binning
fig = histogram(x=data, title="Normal Distribution (Auto-Binning)")
fig.show()
```

## Example 2: Different Normalization Modes

Khisto supports multiple normalization modes: count, percent, probability, and density.

```python
# Count (default)
fig = histogram(x=data, title="Count")
fig.show()
```

## Example 3: Cumulative Distribution

Create cumulative histograms to visualize cumulative distributions.

```python
fig = histogram(x=data, cumulative=True, title="Cumulative Distribution Function")
fig.show()
```

## Example 4: Horizontal Histogram

Orientation can be changed to horizontal by using the `y` parameter or setting `orientation='h'`.

```python
fig = histogram(x=data, orientation="h", title="Horizontal Histogram")
fig.show()
```

## Example 5: Custom Styling

Customize the appearance with templates, opacity, and text labels.

```python
fig = histogram(
    x=data,
    opacity=0.7,
    text_auto=True,
    template="plotly_white",
    title="Styled Histogram",
    width=800,
    height=500,
)
fig.show()
```

## Example 6: DataFrame with Color Grouping

Use DataFrames for more complex visualizations with grouping.

```python
import pandas as pd

# Create a DataFrame with two groups
df = pd.DataFrame(
    {
        "value": np.concatenate(
            [np.random.normal(0, 1, 500), np.random.normal(3, 1.5, 500)]
        ),
        "group": ["A"] * 500 + ["B"] * 500,
    }
)

# Create overlaid histograms
fig = histogram(
    df,
    x="value",
    color="group",
    barmode="overlay",
    opacity=0.6,
    title="Two Groups Comparison",
)
fig.show()
```

```python
# facet histogram
fig = histogram(
    df, x="value", facet_col="group", opacity=0.6, title="Two Groups Comparison"
)
fig.show()
```

## Example 7: Logarithmic Scale

Use log scales for data spanning multiple orders of magnitude.

```python
# Generate exponential data
exp_data = np.random.exponential(1, 1000)

fig = histogram(x=exp_data, log_x=True, title="Exponential Distribution (Log Scale)")
fig.show()
```

## Example 8: Different Distribution Types

See how Khisto adapts to different distribution types.

```python
# Uniform distribution
uniform_data = np.random.uniform(-5, 5, 1000)
fig = histogram(x=uniform_data, title="Uniform Distribution")
fig.show()

# Gamma distribution
gamma_data = np.random.gamma(2, 2, 1000)
fig = histogram(x=gamma_data, title="Gamma Distribution")
fig.show()

# Bimodal distribution
bimodal_data = np.concatenate(
    [np.random.normal(-2, 0.5, 500), np.random.normal(2, 0.5, 500)]
)
fig = histogram(x=bimodal_data, title="Bimodal Distribution")
fig.show()
```

## Example 9: Custom Axis Ranges and Labels

Control axis ranges and customize labels.

```python
fig = histogram(
    x=data,
    range_x=[-4, 4],
    range_y=[0, 1],
    labels={"x": "Measurement (units)", "y": "Frequency"},
    title="Custom Range and Labels",
)
fig.show()
```

## Example 10: Working with Series

Direct support for pandas and polars Series.

```python
import pandas as pd

# Create a pandas Series
series = pd.Series(np.random.normal(5, 2, 1000), name="values")

fig = histogram(x=series, title="From Pandas Series")
fig.show()
```
