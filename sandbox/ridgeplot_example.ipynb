# Ridge Plot Example with Khisto

This notebook demonstrates how to use the `khisto.plot.plotly.ridgeplot` function to visualize multiple distributions with optimal histogram binning. We will generate synthetic data, create a ridge plot, and explore its features.

```python
# Import required libraries

import os

os.environ["KHISTO_BIN_DIR"] = (
    "/home/elouen/python/khisto-python/src/khiops/build/linux-gcc-debug/bin/khisto"
)
import numpy as np
import pandas as pd
from khisto.plot.plotly import ridgeplot
```

## Create Sample Data

Let's generate synthetic data for three categories, each with a different distribution.

```python
# Generate synthetic data for three categories
np.random.seed(42)
n = 500

data = pd.DataFrame(
    {
        "value": np.concatenate(
            [
                np.random.normal(0, 1, n),
                np.random.normal(3, 0.5, n),
                np.random.normal(-2, 0.8, n),
            ]
        ),
        "category": np.repeat(["A", "B", "C"], n),
    }
)
data.head()
```

## Visualize with Ridge Plot

Now let's create a ridge plot to compare the distributions of each category using optimal histogram binning.

```python
# Create the ridge plot
fig = ridgeplot(
    data_frame=data,
    x="value",
    y="category",
    granularity="best",
    title="Ridge Plot of Distributions by Category",
    height=500,
    width=800,
    template="plotly_white",
)
fig.show()
```

## Advanced Example: Temperature Distributions Across Months

Let's create a more compelling example with monthly temperature data showing a clear seasonal trend.

```python
# Generate synthetic temperature data for 12 months
np.random.seed(123)

# Temperature parameters for each month (mean, std)
# Winter months have lower temps, summer months have higher temps
temp_params = {
    "Jan": (5, 3),
    "Feb": (7, 3),
    "Mar": (12, 4),
    "Apr": (16, 4),
    "May": (20, 4),
    "Jun": (25, 3),
    "Jul": (28, 3),
    "Aug": (27, 3),
    "Sep": (23, 4),
    "Oct": (17, 4),
    "Nov": (11, 4),
    "Dec": (6, 3),
}

# Generate temperature data
n_samples = 4000
temp_data = []

for month, (mean, std) in temp_params.items():
    temps = np.random.normal(mean, std, n_samples)
    temp_data.extend([(temp, month) for temp in temps])

# Create DataFrame
df_temps = pd.DataFrame(temp_data, columns=["temperature", "month"])

# Define proper month order
month_order = [
    "Jan",
    "Feb",
    "Mar",
    "Apr",
    "May",
    "Jun",
    "Jul",
    "Aug",
    "Sep",
    "Oct",
    "Nov",
    "Dec",
]

print(f"Generated {len(df_temps)} temperature observations across 12 months")
df_temps.head()
```

```python
# Create ridge plot of temperature distributions
fig_temps = ridgeplot(
    data_frame=df_temps,
    x="temperature",
    y="month",
    category_orders={"month": month_order},
    labels={"temperature": "Temperature (°C)", "month": "Month"},
    title="Temperature Distributions Throughout the Year",
    height=700,
    width=900,
    granularity=6,
    template="plotly_white",
    overlap=0.6,  # More overlap for a denser appearance
    opacity=0.9,
    color_continuous_scale="Viridis",  # Continuous gradient from cold to warm colors
    line_color="auto",  # Automatically use plot background color for lines
)
fig_temps.show()
```
